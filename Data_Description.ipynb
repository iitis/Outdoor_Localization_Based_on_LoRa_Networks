```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
LoRa_Dataset = pd.read_csv('lorawan_antwerp_2019_dataset.csv')
print(LoRa_Dataset)
```

```python
print(np.shape(LoRa_Dataset)) #130429 samples, 77 features ---> 72 gateways (RSSIs), RX Time, SF, HDOP, Latitude and Longitude 
```

```python
LoRa_Dataset.head(5) # Latitude and Longitude are targets
```

```python
print('Minimum Latitude value is', np.min(LoRa_Dataset['Latitude']))
print('Maximum Latitude value is',np.max(LoRa_Dataset['Latitude']))
print('')
print('Minimum Longitude value is', np.min(LoRa_Dataset['Longitude']))
print('Maximum LLongitude value is',np.max(LoRa_Dataset['Longitude']))
```

```python
# Keep only RSSI values for features

LoRa_Dataset = LoRa_Dataset.drop(columns='RX Time')
LoRa_Dataset = LoRa_Dataset.drop(columns='SF')
LoRa_Dataset = LoRa_Dataset.drop(columns='HDOP')
```

```python
# Select all columns starting with "BS" (BS: Gateway)
bs_columns = [col for col in LoRa_Dataset.columns if col.startswith("BS")]
print('There are', len(bs_columns), 'gateways in the dataset.')

# Let's filter the gateways that have -200 in all rows
columns_to_drop = [col for col in bs_columns if (LoRa_Dataset[col] == -200).all()]

print(columns_to_drop)
print(len(columns_to_drop), 'BS have -200 in all rows')
```

```python
#  Remove those gatewaysfrom the dataset
LoRa_Dataset_cleaned = LoRa_Dataset.drop(columns=columns_to_drop)

##  Save the result to a file or review it (optional)
#LoRa_Dataset_cleaned.to_csv("LoRa_Dataset_cleaned.csv", index=False)
```

```python
LoRa_Dataset_cleaned.head(5)
```

```python
LoRa_RSSI = LoRa_Dataset_cleaned.iloc[:, 0:44] # Only RSSI values; datas from gateways
LoRa_RSSI.head(5)
```

* Check Range of RSSI (minimum and maximum RSSI value in the dataset)

```python
df = pd.DataFrame(LoRa_RSSI)

min_value = df.min().min()
max_value = df.max().max()

print(f"Minimum value: {min_value}")
print(f"Maximum value: {max_value}")
```

* To find the minimum RSSI value after -200

```python
# Create DataFrame without the minimum value
filtered_df = df.replace(min_value, np.nan) # Make min_value NaN

# Find the second minimum value
second_min_value = filtered_df.min().min()

print(f"Minimum value: {min_value}")
print(f"Second minimum value: {second_min_value}")
```

* Replacing -200 with the neighboring value, which is -128

```python
df.replace(-200, second_min_value-1, inplace=True)
```

* Check again range of RSSI

```python
min_value = df.min().min()
max_value = df.max().max()

print(f"Minimum value: {min_value}")
print(f"Maximum value: {max_value}")
```
